```python
# matplotlib.pyplotメソッドとnumpyをインポート
import matplotlib.pyplot as plt
import numpy as np

# 6つのグラフを表示する領域を作成
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6))

# 各領域にデータを表示
# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('Line Chart')

# sin関数
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title('Sin Function')

# ヒストグラム
data2 = np.random.randn(1000)
axes[0, 2].hist(data2, bins=30, color='green', alpha=0.7)
axes[0, 2].set_title('Histogram')

# 散布図
x3 = np.random.rand(50)
y3 = np.random.rand(50)
axes[1, 0].scatter(x3, y3)
axes[1, 0].set_title('Scatter Plot')

# その他の2領域は空白

# グラフを表示
plt.tight_layout()  # グラフが重ならないように調整
plt.show()
```
